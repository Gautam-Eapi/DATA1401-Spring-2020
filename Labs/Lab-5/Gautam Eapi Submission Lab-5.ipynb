1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self,max_val):
        self.max_val=max_val
        self.cur_val=1
        
    def increment(self):
        if self.cur_val>self.max_val:
            print("Max value reached.")
        else:
            self.cur_val+=1
            
    def reset(self):
        self.cur_val=1
        
```

3. Implement a class to represent a rectangle, holding the length, width, and $x$ and $y$ coordinates of a corner of the object. Implement functions that compute the area and parameter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class rectangle:
    def __init__(self,length,width,x,y):
        self.__length=length
        self.__width=width
        self.__x=x
        self.__y=y
        
    def area(self):
        return self.__width*self.__length
    
    def perimeter(self):
        return 2*(self.__width+self.__length)
    
    def x(self):
        return self.__x

    def y(self):
        return self.__y
    
    def length(self):
        return self.__length

    def width(self):
        return self.__width
        
```

6. Implement an analogous triangle class.

```python
class triangle:
    def __init__(self,base,height,b,c,x,y):    # the three sides of the triangle are base, b, and c
        self.__base = base
        self.__height = height
        self.__b = b
        self.__c = c
        self.__x = x
        self.__y = y
        
    def area(self):
        return 0.5*self.__base*self.__height
    
    def perimeter(self):
        return self.__base+self.__b+self.__c
    
    def x(self):
        return self.__x

    def y(self):
        return self.__y

    def base(self):
        return self.__base

    def height(self):
        return self.__height

    def b(self):
        return self.__b
    
    def c(self):
        return self.__c
```

```python
tri = triangle(10,4,5,7,0,0)
```

```python
tri.area()
```

```python
tri.perimeter()
```
